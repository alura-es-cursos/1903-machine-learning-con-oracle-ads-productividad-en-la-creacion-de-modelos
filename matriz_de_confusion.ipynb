#Matriz de confusión

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

"""
Infectados por covid-19?

0 -> Não infectado
1 -> Infectado
"""

valores_predichos = [1, 0, 0, 0, 1, 1, 0, 1, 1, 0]
valores_reales = [1, 0, 1, 0, 1, 1, 0, 1, 0, 1]

cm = confusion_matrix(valores_reales, valores_predichos)

group_names = ['True Neg (TN)','False Pos (FP)','False Neg (FN)','True Pos (TP)']
group_counts = [f"{value}" for value in cm.flatten()]
labels = [f"{v1}" for v1 in group_names]
labels = [f"{v1}\n{v2}" for v1, v2 in zip(group_names, group_counts)]

labels = np.asarray(labels).reshape(cm.shape)

ax = plt.subplot()
sns.set(font_scale=1.4) 
sns.heatmap(cm, annot=labels, ax=ax, cmap="viridis", fmt="")

label_font = {'size':'14'}  
ax.set_xlabel('Valores Previstos', fontdict=label_font);
ax.set_ylabel('Valores Reales', fontdict=label_font);

ax.tick_params(axis='both', which='major', labelsize=10)  
ax.xaxis.set_ticklabels(['No infectado', 'Infectado']);
ax.yaxis.set_ticklabels(['No infectado', 'Infectado']);
```
